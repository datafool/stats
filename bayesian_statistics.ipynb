```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draw_beta_distribution(alpha, beta):
    
    # generating 100 x values so as to cover beta distribution probability betweeen 0.01 to 0.99
    x = np.linspace(
        stats.beta.ppf(0.01, alpha, beta),
        stats.beta.ppf(0.99, alpha, beta),
        100
    )
    prob = stats.beta.pdf(x, alpha, beta)
    mean, var, skew, kurtosis = stats.beta.stats(alpha, beta, moments='mvsk')
    print("Mean {}, Variance {}, Skew {}, Kurtosis {}".format(mean, var, skew, kurtosis))
    
    max_prob_arg = np.argmax(prob)
    max_prob = prob[max_prob_arg]
    max_x = x[max_prob_arg]
    min_prob = np.min(prob)

    plt.figure(figsize=(20,5))
    plt.plot(x, prob, '-o')
    plt.vlines(max_x, min_prob, max_prob)
    plt.xlabel("theta")
    plt.ylabel("prob density")
    plt.title("Beta distribution for alpha {} and beta {}".format(alpha, beta))
    
```

```python
def draw_prioir_posterior_likelihood(n, k, alpha, beta,theta_hyposthesis, limit_y=True):
    x = np.linspace(0, 1,100)
    prior = stats.beta.pdf(x, alpha, beta)
#     likelihood = stats.binom.pmf(k, n, x)
    # We use beta distirbution for likelihood as binomial can be approximated as beta distirbution
    # However, this question still needs to be answered and discussed
    likelihood_beta = stats.beta.pdf(x, k+1, n -k + 1)
#     import pdb; pdb.set_trace()
    alpha_posterior = alpha + k + 1 - 1
    beta_posterior = beta + (n - k) + 1 - 1
    posterior_via_beta = stats.beta.pdf(x, alpha_posterior, beta_posterior)
#     posterior_via_multiplication = np.multiply(prior, likelihood)
#     posterior_via_multiplication = posterior_via_multiplication
    #credible interval
    ci_lower = stats.beta.ppf(0.025, alpha_posterior, beta_posterior)
    ci_upper = stats.beta.ppf(0.975, alpha_posterior, beta_posterior)
    print("95% CI Interval ({}, {})".format(ci_lower, ci_upper))
    prior_val = stats.beta.pdf(theta_hyposthesis, alpha, beta)
    posterior_val = stats.beta.pdf(theta_hyposthesis, alpha_posterior, beta_posterior)
    bayes_factor = posterior_val / prior_val 
                    
    print("Bayes Factor for fair coin is {}".format(bayes_factor))
    
    posterior_mean = alpha_posterior / (alpha_posterior + beta_posterior)
    print("Posterior mean is {}".format(posterior_mean))
    plt.figure(figsize=(20,5))
    plt.plot(x, prior, label='prior')
#     plt.plot(x, likelihood, label='binomial likelihood')
    plt.plot(x, likelihood_beta, '--', label='beta likelihood')
    plt.plot(x, posterior_via_beta, '*', label='posterior_via_beta')
    plt.vlines(ci_lower, 
               np.min(posterior_via_beta), 
               np.max(posterior_via_beta), 
               color='r', 
               linestyles='dashed',
               label='95% ci_lower')
    
    plt.vlines(ci_upper, 
               np.min(posterior_via_beta), 
               np.max(posterior_via_beta), 
               color='r', 
               linestyles='dashed',
               label='95% ci_upper')
    plt.vlines(theta_hyposthesis, prior_val, posterior_val, label='bayes factor fair coin')
    plt.vlines(posterior_mean, 
               np.min(posterior_via_beta), 
               np.max(posterior_via_beta),
               linestyle='dashed',
               color='g',
               label='posterior mean')
    
    if limit_y:
        plt.ylim(0,10)
#     plt.plot(x, posterior_via_multiplication, label='posterior_via_mul')
    plt.legend()
    plt.title("Prior, Likelihood and Posterior for alpha={}, beta={}, n={}, k={}".format(alpha, beta, n, k))
```

```python
draw_prioir_posterior_likelihood(20, 10, 1, 1, 0.5)
```

posterior alpha = prior alpha + k
posterior beta = prior beta + (n - k)

**Note**:: 
- A binomial distribution can be approximated to beta distribution with alpha = k + 1 and beta = n - k +1
- So, we can take likelihood as beta distirbution with alpha and beta as derived above
    - But beta distribution thus derived does not match with binomial distribution
